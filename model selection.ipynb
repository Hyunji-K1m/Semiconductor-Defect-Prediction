```python
corr_df = df.corr()

corr_df
```

```python
df_nan = corr_df[corr_df > abs(0.8)]

df_nan
```

```python
col_names = list(df_nan)
row_names = list(df_nan.index)
```

```python
corr_result = []
for i in range(0, len(col_names)):
    for j in range(0,len(row_names)):
        value = []
        if ( (df_nan[col_names[i]][row_names[j]] > 0.8) & (df_nan[col_names[i]][row_names[j]] < 1 )) :
            value.append(col_names[i])
            value.append(row_names[j])
            value.append(df_nan[col_names[i]][row_names[j]])
            corr_result.append(value)
```

```python
corr_result
```

```python
corr_result=pd.DataFrame(corr_result)
corr_result
```

```python
corr_result=corr_result.drop_duplicates([2], keep="first")
```

```python
corr_result
```

```python
a = corr_result[0].value_counts()
a0=pd.DataFrame(a)
b = corr_result[1].value_counts()
b0=pd.DataFrame(b)
c = pd.concat([a0,b0], ignore_index=True, axis=1)
c
```

```python
c = c.fillna(0)
c['sum'] = c[0]+c[1]

c
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from tqdm import tqdm_notebook
import warnings
warnings.filterwarnings(action='ignore')
```

```python
df=pd.read_csv("secom_ffinal.csv")
df.drop("Unnamed: 0",axis=1,inplace=True)
```

```python
df
```

```python
X = df.drop("Pass/Fail", axis=1)
y = df["Pass/Fail"].values
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
RS = RobustScaler()
RS.fit(X)
X_rs = RS.transform(X)
```

```python
X_rs_pd = pd.DataFrame(X_rs, columns=X.columns)
X3 = X_rs_pd
```

```python
feature_name = list(X3.columns)
feature_name
```

## RandomForest

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier
```

```python
embeded_rf_selector = SelectFromModel(RandomForestClassifier(n_estimators=100))
```

```python
embeded_rf_selector.fit(X3, y)
```

```python
embeded_rf_support = embeded_rf_selector.get_support()
embeded_rf_feature = X3.loc[:,embeded_rf_support].columns.tolist()
print(str(len(embeded_rf_feature)), 'selected features')
```

```python
df_rf = pd.DataFrame(X3, columns=embeded_rf_feature)
X_rf = df_rf.copy() 
```

```python
X_rf
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
embeded_lr_selector = SelectFromModel(LogisticRegression(penalty="l1",solver='liblinear'))
```

```python
embeded_lr_selector.fit(X3, y)
```

```python
embeded_lr_support = embeded_lr_selector.get_support()
embeded_lr_feature = X3.loc[:,embeded_lr_support].columns.tolist()
print(str(len(embeded_lr_feature)), 'selected features')
```

```python
df_lr = pd.DataFrame(X3, columns=embeded_lr_feature)
X_lr = df_lr.copy() 
```

```python
X_lr
```

## LGBM

```python
from lightgbm import LGBMClassifier

lgbc=LGBMClassifier(n_estimators=500, learning_rate=0.05, num_leaves=32, colsample_bytree=0.2,
            reg_alpha=3, reg_lambda=1, min_split_gain=0.01, min_child_weight=40)

embeded_lgb_selector = SelectFromModel(lgbc)
embeded_lgb_selector.fit(X3, y)
```

```python
embeded_lgb_support = embeded_lgb_selector.get_support()
embeded_lgb_feature = X3.loc[:,embeded_lgb_support].columns.tolist()
print(str(len(embeded_lgb_feature)), 'selected features')
```

```python
df_lgb = pd.DataFrame(X3, columns=embeded_lgb_feature)
X_lgb = df_lgb.copy() 
```
